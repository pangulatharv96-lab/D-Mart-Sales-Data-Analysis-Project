<h4>For Numeric_columns 

```python
data["Price"]=data["Price"].fillna(data["Price"].mean())
```

```python
data["DiscountedPrice"]=data["DiscountedPrice"].fillna(data["DiscountedPrice"].mean())
```

```python
data.isnull().sum()
```

```python
data['Name'].fillna('Unknown', inplace=True)
data['Brand'].fillna('Unknown', inplace=True)
data['Category'].fillna('Other', inplace=True)
data['SubCategory'].fillna('Other', inplace=True)
data['Quantity'].fillna('Unknown', inplace=True)
```

```python
plt.plot(data['Price'], label='Price')
plt.plot(data['DiscountedPrice'], label='Discounted Price')

plt.xlabel('Index')
plt.ylabel('Amount')
plt.title('Price vs Discounted Price')
plt.legend()
plt.show()
```

```python
top_brands = data['Brand'].value_counts().head(10)

top_brands.plot(kind='bar')
plt.xlabel('Brand')
plt.ylabel('Number of Products')
plt.title('Top 10 Brands')
plt.show()
```

```python
plt.hist(data['Price'], bins=30)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Price Distribution')
plt.show()
```

```python
data['DiscountPercent'] = ((data['Price'] - data['DiscountedPrice']) / data['Price']) * 100

plt.hist(data['DiscountPercent'], bins=30)
plt.xlabel('Discount %')
plt.ylabel('Frequency')
plt.title('Discount Percentage Distribution')
plt.show()
```
